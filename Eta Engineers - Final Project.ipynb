# **1. Business Problem Understanding**
****

**Context**

Sebuah perusahaan Asuransi Kendaraan di AS ingin meningkatkan revenue. Ada banyak faktor yang dapat memengaruhi kesuksesan sebuah perusahaan Asuransi. Pelanggan atau customer adalah salah satu faktor utama yang menentukan kesuksesaan perusahaan. Namun mereka menghadapi masalah dalam marketingnya. Mereka ingin mempertahankan pelanggannya dan juga menarik pelanggan baru. 

Perusahaan sudah melakukan evaluasi, menawarkan perpanjangan polis dan mendapatkan respon dari customer. Namun itu belum cukup untuk menaikkan revenue perusahaan. Maka dari itu, mereka mempertimbangkan CLV (Customer Lifetime Value) sebagai parameter untuk tujuan ini.

CLV berarti nilai umur pelanggan. Apa yang dimaksud dengan umur pelanggan di sini bukanlah umur pelanggan dilihat dari tahun kelahirannya, melainkan dari rentang waktu ia menggunakan produk asuransi kendaraan tersebut. Lebih jelas lagi, CLV adalah sebuah metrik untuk memperkirakan total nilai pelanggan terhadap sebuah perusahaan asuransi dalam jangka waktu tertentu (mengikuti durasi hubungan bisnis keduanya). Secara sederhana, CLV adalah prediksi atas nilai total pendapatan yang bisa didapatkan dari pelanggan tersebut.

**Problem Statement :** <br>

  Penerapan strategi atau treatment terdapat setiap pelanggan dapat berbeda- beda. Apabila dilakukan treatment yang sama ke seluruh pelanggan akan memakan waktu dan sumber daya yang lebih, terlebih lagi treatment yang tidak tepat sasaran dapat menyebabkan berkurangnya keuntungan hingga mengalami kerugian bagi perusahaan.

  Sebagai contoh, pelanggan baru membutuhkan promosi dan iklan dimana salah satunya berupa potongan harga atau pengenalan jasa/ asuransi jenis baru, sedangkan untuk pelanggan lama, salah satu treatmentnya adalah dengan memberikan reward kepada mereka sehingga pelanggan tersebut tidak akan beralih ke kompetitor. 
  Secara umum, customer lifetime value ini akan membantu perusahaan untuk : 
  1. Mempertahankan loyal customer dan akuisisi pelanggan
  2. Strategi marketing yang akan diterapkan terhadap customer berdasarkan CLV


Dalam masalah yang melibatkan banyak variabel ini, dapat menjadi tantangan untuk memutuskan feature mana yang harus dimasukkan ke dalam model. Beberapa pendekatan yang digunakan dalam model ini diharapkan dapat menggambarkan hubungan antara CLV sebagai target variable dan feature yang telah diberikan di dalam data perusahaan. Kemudian, dalam melakukan prediksi terhadap CLV kita dapat gunakan dengan mengajukan pertanyaan, diantaranya:

1. Bagaimana memprediksi CLV, sehingga dapat meningkatkan revenue perusahaan?

**Encoding**
****

Sekarang kita melakukan fitur encoding untuk fitur-fitur categorical yang kita miliki.
Yang akan kita lakukan adalah :


1. Mengubah fitur/kolom `State`, `Coverage`, `EmploymentStatus`, `Gender`, `Location Code`, `Marital Status`, `Policy Type`, `Renew Offer Type`, `Sales Channel` dan `Vehicle Size` menggunakan One Hot Encoding, karena fitur ini tidak memiliki urutan/tidak ordinal, dan juga jumlah unique datanya hanya sedikit.
2. Mengubah fitur/kolom `Education` menggunakan Ordinal Encoding, karena fitur ini memiliki urutan atau  ordinal, dan juga jumlah unique datanya cukup banyak. 

3. Mengubah fitur/kolom `Policy` dan `Vehicle Class` menggunakan Binary Encoding, karena fitur ini memiliki unique data yang banyak dan tidak memiliki urutan/tidak ordinal, bila kita menggunakan One Hot Encoding akan terlalu banyak fitur baru yg terbuat, dan apabila kita menggunakan Ordinal/Label Encoding hasilnya dapat kurang cocok/kurang baik. Oleh karena itu kita akan mencoba menggunakan Binary Encoding saja.


Kita melakukan Scaling pada fitur `Income` dan `Total Claim Amoint` menggunakan Robust Scaler. hal ini dilakukan karena kedua fitur tersebut memiliki range data yang cukup jauh dan data memiliki data outlier.

Sedangkan untuk fitur `Monthly Premium Auto`, `Months Since Last Claim`, `Months Since Policy Inception`, `Number of Open Complaints`, `Number of Open Complaints` dan `Number of Policies` tidak dilakukan proses Scaling Karena semua data ini tidak memili range data yang terlalu jauh.

```python
ordinal_mapping  = [{
    'col':'Education',
    'mapping' : {'High School or Below':0, 'College':1, 'Bachelor':2, 'Master':3, 'Doctor':4}
}]
```

Proses Maping pada Fitur `Education`.

```python
# Encoding 
# Mengubah fitur kategorikal menjadi numerikal
transformer = ColumnTransformer([
                                 ('one hot encoder',OneHotEncoder(drop = 'first'),['State', 'Response', 'Coverage',  'Gender', 
                                 'Location Code', 'Marital Status', 'Policy Type','Sales Channel', 'Vehicle Size']),
                                 ('ordinal encoding', ce.OrdinalEncoder(cols=['Education'], mapping=ordinal_mapping), ['Education']),
                                 ('binary encoder',ce.BinaryEncoder(),['Vehicle Class']),
                                 ('robust scaler',RobustScaler(),['Income','Total Claim Amount'])
                                ], remainder = 'passthrough')
```

```python
transformer_rf = ColumnTransformer([
                                 ('one hot encoder',OneHotEncoder(drop = 'first'),['State', 'Response', 'Coverage',  'Gender', 
                                 'Location Code', 'Marital Status', 'Policy Type','Sales Channel', 'Vehicle Size']),
                                 ('ordinal encoding', ce.OrdinalEncoder(cols=['Education'], mapping=ordinal_mapping), ['Education']),
                                 ('binary encoder',ce.BinaryEncoder(),['Vehicle Class'])
                                ], remainder = 'passthrough')
```

Karena Model Randomforest merupakan tree based model maka proses scaling pada data tidak memberikan dampak pada proses modeling. sehingga kita tidak melakukan proses scaling.

# **6. Modelling**
****

```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=42)
rf = RandomForestRegressor(random_state=42)
xgb = XGBRegressor(random_state=42)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)
```

```python
# Kandidat algoritma
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]


score_r2 = []
nilai_mean_r2 = []
nilai_std_r2 = []

# Mencari algoritma terbaik berdasarkan nilai R-Square
for i in models:
    
    skfold = KFold(n_splits=5, shuffle=True, random_state=42)
    estimator = Pipeline([
        ('preprocessing', transformer),
        ('model', i)
    ])

    # R-Square
    model_cv_r2 = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=skfold, 
        scoring='r2'
        #error_score='raise'
        )
    #print(model_cv_r2, i)
    nilai_mean_r2.append(model_cv_r2.mean())
    nilai_std_r2.append(model_cv_r2.std())
```

```python
# Evaluasi hasil dari 5 kandidat algoritma yang digunakan
pd.DataFrame({
    'Model': ['Linear Regression', 'KNN Regressor', 'DecisionTree Regressor', 'RandomForest Regressor', 'XGBoost Regressor'],
    'Mean_R2': nilai_mean_r2,
    'Std_R2' : nilai_std_r2
}).set_index('Model').sort_values(by='Mean_R2', ascending=False)
```

**Model Benchmarking : Test Data**
****

```python
# Prediksi pada test set
score_r2 = []

models = {
    'XGB': XGBRegressor(random_state=42),
    'RandomForest': RandomForestRegressor(random_state=42),
    'LinearRegression' : LinearRegression(),
    'Knn' : KNeighborsRegressor(),
    'DecisionTree' : DecisionTreeRegressor(random_state=42)
}


for i in models:

    model = Pipeline([
        ('preprocessing', transformer),
        ('model', models[i])
        ])

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score_r2.append(r2_score(y_test, y_pred))
```

```python
pd.DataFrame({
    'Model': ['Linear Regression', 'KNN Regressor', 'DecisionTree Regressor', 'RandomForest Regressor', 'XGBoost Regressor'],
    'R2': nilai_mean_r2,
}).set_index('Model').sort_values(by='R2', ascending=False)
```

Terlihat kembali lagi model Randomforest Regressor adalah yang terbaik performanya pada test data. Sehingga Model Randomforest Regressor lah yang akan kita gunakan pada model kali ini.

### **Hyperparameter Tuning**
****

```python
hyperparam_space= {
    'model__n_estimators': [100, 125, 150, 175, 200, 225, 250],
    'model__criterion':['squared_error'],
    'model__min_samples_split': [1, 2, 3, 4, 5, 6],
    'model__min_samples_leaf' : [1, 2, 3, 4, 5, 6 , 7, 8]
}
# , 'absolute_error', 'poisson'
```

Default Hyperparameter Randomforest Regressor => `n_estimator` = 100, `criterion` = 'squared_error', `min_samples_split` = 2, `min_samples_leaf` = 1

```python
# Benchmark model dengan hyperparameter tuning

# Membuat algorithm chains
estimator_rf = Pipeline([
        ('preprocessing', transformer_rf),
        ('model', rf)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=42)

# Hyperparameter tuning
random_rf = RandomizedSearchCV(
    estimator_rf, 
    param_distributions = hyperparam_space,
    n_iter = 75,
    cv = crossval, 
    scoring = 'r2', 
    n_jobs = -1,
    random_state = 42
)
```

```python
# Fitting data training untuk mencari parameter terbaik
random_rf.fit(X_train, y_train)
```

```python
# Cek skor dan params terbaik
print('Random Forest')
print('Best_score:', random_rf.best_score_)
print('Best_params:', random_rf.best_params_)
```

Hyperparameter terbaik Randomforest Regressor dari hasil RandomSearchCv => `n_estimators` = 200, `Criterion` = 'squared_error', `min_samples_split` = 5, `min_samples_leaf` = 4.

Untuk hyperparameter `Criterion` default valuenya tetap yang terbaik, untuk `n_estimators` yang optimal dari randomsearch kali ini adalah 200 dari nilai default 100, untuk `min_samples_split` yang optimal adalah 5 dari nilai default 2 dan untuk `min_samples_leaf` yang optimal adalah 4 dari nilai default 1.

**Perbandingan Score Sebelum dan Sesudah Tuning**

```python
# Define model terhadap estimator terbaik
rf_tuning = random_rf.best_estimator_

# Fitting model
rf_tuning.fit(X_train, y_train)
```

```python
# Membuat algorithm chains
estimator_default = Pipeline([
        ('preprocessing', transformer_rf),
        ('model', rf)
        ])
estimator_default.fit(X_train, y_train)
```

```python
# predict default model
y_pred_default = estimator_default.predict(X_test)
r2_default=r2_score(y_test, y_pred_default)

# predict tuned model
y_pred_rf_tuning = rf_tuning.predict(X_test)
r2_tuned=r2_score(y_test, y_pred_rf_tuning)

print('R-Square Score default Randomforest', r2_default)
print('R-Square Score Tuned Randomforest', r2_tuned)
```
